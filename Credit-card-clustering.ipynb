# Data Preprocessing

```python
df.drop('CUST_ID', axis=1, inplace=True)
```

```python
df.info()
```

Standardize data

```python
# Import StandardScaler classifier
from sklearn.preprocessing import StandardScaler
```

```python
# Initialize the classifier: scaler
scaler = StandardScaler()
```

```python
# fit and transform scaler into the df
df_scaled = scaler.fit_transform(df)
```

PCA

```python
# Import PCA 
from sklearn.decomposition import PCA
```

```python
# Initialize PCA
pca = PCA()
```

```python
# fit PCA to the scaled data
pca.fit(df_scaled)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.show()
```

```python
# From the PCA, the intrisic dimension is 2
n_components = 2

# Execute PCA
pca = PCA(n_components=n_components)

# fit and transform pca into df
df_pca = pca.fit_transform(df_scaled)
```

Finding the number of k cluster

```python
# Import KMeans module
from sklearn.cluster import KMeans
```

```python
ks = range(1, 10)
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters = k)
    
    # Fit model to samples
    model.fit(df_pca)
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
# From the diagram above, the best number of cluster, k is 4
n_clusters = 3

# Initializing KMeans with the optimal number of clusters 
kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# fit the kmeans to df
kmeans.fit(df_pca)

labels = kmeans.predict(df_pca)

# Visualize clusters using the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=labels, cmap='viridis', marker='o', edgecolors='black')
plt.title('KMeans Clustering')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
df_imputed = pd.DataFrame(df_scaled, columns=df.columns)
type(df_imputed)
```

```python
best_features = ['BALANCE','PURCHASES','PAYMENTS']

# dataframe with best columns
df_final = pd.DataFrame(df_imputed[best_features])

print('New dataframe with best columns has just been created. Data shape: ' + str(df_final.shape))
```

```python
# apply KMeans clustering
alg = KMeans(n_clusters = 3)
label = alg.fit_predict(df_final)
```

```python
# adding label columns
df_final['label'] = kmeans.labels_

# create statistical table sort by label and calculate the mean
stat_credit = df_final.groupby('label').mean()
print(stat_credit)
```

```python
# create a 'cluster' column
df_final['cluster'] = label
best_features.append('cluster')
```

```python
# make a Seaborn pairplot
plt.figure(figsize=(15,8))
sns.pairplot(df_final[best_features], hue='cluster', palette='coolwarm')
```

The first cluster (Cluster 0) has low balance, low purchases, and low payments. This cluster might need financial management advice in order to help them have a better way for their financial status.

Second cluster (Cluster 1) has a varied balance, high purchases, and high payments. This cluster is already in an ideal state, but some of them have low balance so we might give them some advice on financial management too.

The third cluster (Cluster 2) has high balance, low purchases, and low payments. The last cluster is the focus target among all the clusters as their account balance is very high but very low in purchases and payment.
